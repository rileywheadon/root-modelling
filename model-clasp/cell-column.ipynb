```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from numba import njit
from modules import get_mutant_data
```

```python
%matplotlib inline
mpl.rcParams['font.size'] = 18
PATH = "../summary/img"
```

```python
# 0 < C_WT < C_BC
# R_C1 < R_WT < R_BC
PARAMS = (
    0.02500,    # g_0
    0.01400,    # C_WT
    0.02200,    # C_BC
    0.00028,    # R_C1
    0.00029,    # R_WT
    0.00030,    # R_BC
    0.05000,    # d_0
    20.0000,    # d_L
    20.0000,    # n
)

STEP = 0.01
MAX_CELLS = 1000
MAX_TIME = 500
MIN_SIZE = 9
MAX_SIZE = 150

VT = np.arange(0, MAX_TIME + STEP, STEP) 
```

```python
@njit
def simulate_step(mparams, L, P, D, i):

    # Unpack the parameters, create lambda functions
    G_0, G_1, d_0, d_L, n = mparams
    dL = lambda l, p: (G_0 + (G_1 * p)) * l * STEP
    dD = lambda l: d_0 * (1 - ((l ** n) / ((d_L ** n) + (l ** n)))) * STEP
    
    # Unpack the data from the previous and current row
    L0, P0, D0 = L[i-1, :], P[i-1, :], D[i-1, :]
    L1, D1 = L[i, :], D[i, :]

    # Iterate through the (i-1)-th row and update
    j, k = 0, 0
    while L0[j] > 0:

        # Handle the division case
        if D0[j] >= 1 and L0[j] > MIN_SIZE:
            D1[k], D1[k+1] = 0, 0
            L1[k], L1[k+1] = L0[j]/2, L0[j]/2
            k += 1

        # Handle the growth case
        elif L0[j] < MAX_SIZE:
            D1[k] = D0[j] + dD(L0[j])
            L1[k] = L0[j] + dL(L0[j], P0[j])

        # Handle the differentiation case
        else:
            D1[k] = D0[j] + dD(L0[j])
            L1[k] = L0[j]
        
        k += 1
        j += 1

    # Update the position vector and return
    return L1, np.cumsum(L1), D1
            
        
def simulate_root(params, vt, max_cells, mutant):

    # Modify the parameters based on the mutant
    g_0, C_BC, C_WT, R_C1, R_BC, R_WT, d_0, d_L, n= params

    if mutant == "C1":
        G_0 = g_0
        G_1 = R_C1
    if mutant == "BC":
        G_0 = g_0 - C_BC
        G_1 = R_BC
    if mutant == "WT":
        G_0 = g_0 - C_WT
        G_1 = R_WT
    
    mparams = (G_0, G_1, d_0, d_L, n)

    # Create arrays for cell lengths, positions, and division statuses
    L = np.zeros((vt.size, max_cells)) 
    P = np.zeros((vt.size, max_cells)) 
    D = np.zeros((vt.size, max_cells)) 

    # Set the initial lengths and positions
    L[0, :10] = [7, 7, 7, 7, 7, 7, 7, 7, 7, 7]
    P[0, :] = np.cumsum(L[0, :])

    # Run the simulation loop and return
    for i in np.arange(1, vt.size):
        L[i, :], P[i, :], D[i, :] = simulate_step(mparams, L, P, D, i)

    return L, P, D
```

```python
WT = simulate_root(PARAMS, VT, MAX_CELLS, "WT")
BC = simulate_root(PARAMS, VT, MAX_CELLS, "BC")
C1 = simulate_root(PARAMS, VT, MAX_CELLS, "C1")
```

```python
def process_data(data):

    # Filter the data to include points from every 500 time steps
    L, P, D = data
    L = L[np.arange(VT.size) % 500 == 0, :]
    P = P[np.arange(VT.size) % 500 == 0, :]

    # Flatten the L and P arrays
    L = L.flatten()
    P = P.flatten()

    # Create a new data array containing only nonzero lengths
    data = np.stack((P, L), axis = 1)
    data = data[(data[:, 1] > 0) & (data[:, 1] < MAX_SIZE)]
    return data

def plot_position_length(WT, BC, C1):
    mpl.rcParams['figure.figsize'] = (10, 10)

    # Filter the data to only include points from every 100 time steps
    WT = process_data(WT)
    BC = process_data(BC)
    C1 = process_data(C1)

    # Create a scatterplot
    plt.xlabel(r"Position ($\mu$m)")
    plt.ylabel(r"Length ($\mu$m)")
    plt.title("Position vs. Length")
    plt.xlim((0, 1000))
    plt.ylim((0, 150))
    plt.yticks(np.arange(0, 180, 30))
    plt.scatter(WT[:,0], WT[:, 1], color = "c", alpha = 0.2, edgecolor = "none", label = "WT Model")
    plt.scatter(BC[:,0], BC[:, 1], color = "k", alpha = 0.2, edgecolor = "none", label = "BC Model")
    plt.scatter(C1[:,0], C1[:, 1], color = "r", alpha = 0.2, edgecolor = "none", label = "C1 Model")

    # Get the fitted functions to the observed data for comparison
    wt_data, wt_fit = get_mutant_data("trichoblast", "WT-", 10, 12)
    bc_data, bc_fit = get_mutant_data("trichoblast", "BC-", 10, 12)
    c1_data, c1_fit = get_mutant_data("trichoblast", "C1-", 10, 12)

    vP = np.linspace(0, 1000, 1001)
    plt.plot(vP, wt_fit(vP / 1000), "c", lw = 3, label = "WT Data Avg.")
    plt.plot(vP, bc_fit(vP / 1000), "k", lw = 3, label = "BC Data Avg.")
    plt.plot(vP, c1_fit(vP / 1000), "r", lw = 3, label = "C1 Data Avg.")
    plt.legend()

    plt.savefig(f"{PATH}/column-position-length.png")

plot_position_length(WT, BC, C1)
```

```python
def find_last_division(data):
    L, P, D = data
    divisions = P[np.where(D > 1)]
    return np.max(divisions)
    
def plot_division_zone(WT, BC, C1):
    mpl.rcParams['figure.figsize'] = (10, 15)
    
    fig, (ax1, ax2, ax3) = plt.subplots(nrows = 3, ncols = 1, sharex = True, layout = "constrained")
    ax1.set_xlim((300, 400))
    
    ax1.set_title(rf"Wild Type ({np.max(WT[1]):.0f}$\mu$m)")
    ax1.set_ylim((0, 400))
    wt_last = find_last_division(WT)
    wt_div = ax1.hlines(wt_last, 0, MAX_TIME, colors = "k", linestyles = "dashed", lw = 3)

    ax2.set_title(rf"BRIN-CLASP ({np.max(BC[1]):.0f}$\mu$m)")
    ax2.set_ylim((0, 400))
    bc_last = find_last_division(BC)
    bc_div = ax2.hlines(bc_last, 0, MAX_TIME, colors = "k", linestyles = "dashed", lw = 3)

    ax3.set_title(rf"CLASP-1 ({np.max(C1[1]):.0f}$\mu$m)")
    ax3.set_ylim((0, 400))
    c1_last = find_last_division(C1)
    c1_div = ax3.hlines(c1_last, 0, MAX_TIME, colors = "k", linestyles = "dashed", lw = 3)

    for i in range(MAX_CELLS):
        ax1.scatter(VT, WT[1][:, i], color = "g", s = 0.2)
        ax2.scatter(VT, BC[1][:, i], color = "g", s = 0.2)
        ax3.scatter(VT, C1[1][:, i], color = "g", s = 0.2)      

    fig.legend([wt_div], ["Last Division"], loc='outside upper left')
    fig.supxlabel(r"Time")
    fig.supylabel(r"Position ($\mu$m)")
    fig.suptitle("Division Zone Model", fontweight = "bold")

    fig.savefig(f"{PATH}/column-division-zone.png")

plot_division_zone(WT, BC, C1)
```

```python
def get_divisions(data):
    L, P, D = data
    return P[np.where((D > 1) & (L > MIN_SIZE))]

def plot_division_locations(WT, BC, C1):
    mpl.rcParams['figure.figsize'] = (10, 12)
    wt_divs = get_divisions(WT)
    bc_divs = get_divisions(BC)
    c1_divs = get_divisions(C1)

    print(np.mean(wt_divs), np.median(wt_divs), np.max(wt_divs), np.size(wt_divs))
    print(np.mean(bc_divs), np.median(bc_divs), np.max(bc_divs), np.size(bc_divs))
    print(np.mean(c1_divs), np.median(c1_divs), np.max(c1_divs), np.size(c1_divs))

    fig, (ax1, ax2, ax3) = plt.subplots(nrows = 3, ncols = 1, sharex = True, layout = "constrained")
    fig.suptitle("Histogram of Division Positions", fontweight = "bold")
    fig.supxlabel(r"Position ($\mu$m)")
    fig.supylabel("Count")
    
    ax1.set_ylim((0, 80))
    ax2.set_ylim((0, 80))
    ax3.set_ylim((0, 80))
    
    ax1.hist(wt_divs, bins = 16, range=(0, 400), color = 'c', alpha = 0.5)
    ax2.hist(bc_divs, bins = 16, range=(0, 400), color = 'k', alpha = 0.5)
    ax3.hist(c1_divs, bins = 16, range=(0, 400), color = 'r', alpha = 0.5)

    ax1.set_title("Wild Type")
    ax2.set_title("BRIN-CLASP")
    ax3.set_title("CLASP-1")

    fig.savefig(f"{PATH}/column-division-histogram.png")

plot_division_locations(WT, BC, C1)
```

```python
def plot_comparison_setup(model, data, name, fname):

    plt.xlabel(r"Position ($\mu$m)")
    plt.ylabel(r"Length ($\mu$m)")
    plt.title(f"{name} Model vs. Observations")
    plt.xlim((0, 1000))
    plt.ylim((0, 150))
    plt.yticks(np.arange(0, 180, 30))

    plt.scatter(model[:,0], model[:, 1], color = "r", alpha = 0.4, edgecolor = "none", label = f"{name} Model")
    plt.scatter(data[:, 0], data[:, 1], color = "b", alpha = 0.4, marker = "x", label = f"{name} Observations")
    
    plt.savefig(f"{PATH}/column-{fname}.png")
    plt.legend()
    plt.show()
    

def plot_comparison(WT, BC, C1):
    mpl.rcParams['figure.figsize'] = (10, 10)

    # Filter the data to only include points from every 500 time steps
    WT = process_data(WT)
    BC = process_data(BC)
    C1 = process_data(C1)

    # Get the observed data for comparison
    wt_data, wt_fit = get_mutant_data("trichoblast", "WT-", 10, 12)
    bc_data, bc_fit = get_mutant_data("trichoblast", "BC-", 10, 12)
    c1_data, c1_fit = get_mutant_data("trichoblast", "C1-", 10, 12)

    # Create the scatterplots
    plot_comparison_setup(WT, wt_data, "Wild Type", "wild-type")
    plot_comparison_setup(BC, bc_data, "BRIN-CLASP", "brin-clasp")
    plot_comparison_setup(C1, c1_data, "CLASP-1", "clasp-1")

plot_comparison(WT, BC, C1)
```
