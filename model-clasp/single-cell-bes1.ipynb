```python
# Define the time vector
STEP = 0.01
vT = np.arange(0, 24 + STEP, STEP)

# Load in data sets
POSITION, time_to_position, position_to_time = get_position()
BES1 = get_bes1_data()
BL = get_bl(np.linspace(0, 1001, 1000), 0.5, 50)

# Transform the BES1 data to be in terms of time
BES1[:, 0] = position_to_time(BES1[:, 0])

# Get the vB vector which is transformed into time
vB = np.interp(time_to_position(vT), BL[:, 0], BL[:, 1])
```

```python
# Function for computing the bound receptors
@njit
def RB(B, R_T, K_D):
    A = B + R_T + K_D
    return (A - np.sqrt(A ** 2 - 4 * R_T * B)) / 2

# Run a complete simulation
@njit
def simulate_cell(data, params):

    s_0, s_in, s_out = params
    vRB = RB(vB, 62, 10)
    s, vS = s_0, np.empty(vRB.size)
    
    for i in prange(vRB.size):
        vS[i] = s
        s += (s_in * vRB[i] - s_out * s) * STEP

    predicted = np.interp(BES1[:, 0], vT, vS)
    rmse = RMSE(predicted, BES1[:, 1])
    return vS, rmse, predicted
```

## Model Best Fit

```python
def fit_model(data, log):

    # Set initial parameter values and bounds
    params = np.array([0.25, 0.25, 0.25])
    bounds = np.array([[0, 0.5], [0, 0.5], [0, 0.5]])

    # Create a cost function
    cost = lambda params : simulate_cell(data, params)[1]

    # Find the parameters of best fit
    fit = minimize(
        cost, 
        params, 
        method = "trust-constr",
        bounds = bounds, 
        options = {"maxiter": 1000}
    )

    # Run a simulation with the optimal parameters
    vS, rmse, predicted = simulate_cell(BES1[:, 1], fit.x)
    s_0, s_in, s_out = fit.x

    # Log the simulation
    if log: 
        print("Success: ", fit.success, fit.message)
        print(f"Params: {s_0:.2e}, {s_in:.2e}, {s_out:.2e}")
        print("Error: ", rmse)
        
    return vS, rmse, predicted, fit.x

vS, rmse, predicted, optimal = fit_model(BES1[:, 1], True)
```

```python
# Configure the plot
fig, ax = plt.subplots(nrows = 1, ncols = 1)
ax.set_title("Fitted BES1 Signalling Models")
ax.set_xlim((0, 24))
ax.set_xticks(np.arange(0, 28, 4))
ax.set_xlabel(r"Time ($h$)")
ax.set_ylabel("BES1 Transcription Factor (a.u.)")

# Fit a linear curve to the data
f = lambda x, A, B : A + B * x
linear_fit, linear_rmse = fit_curve(f, BES1)
linear_aic = AIC(3, linear_fit(BES1[:, 0]), BES1[:, 1])
ax.plot(vT, linear_fit(vT), color = "limegreen", lw = 3, label = f"(RMSE = {linear_rmse:.4f})")

# Fit an exponential to the data
g = lambda x, A, B, C : A + B * np.exp(C * x)
exp_fit, exp_rmse = fit_curve(g, BES1)
exp_aic = AIC(4, linear_fit(BES1[:, 0]), BES1[:, 1])
ax.plot(vT, exp_fit(vT), color = "slateblue", lw = 3, label = f"(RMSE = {exp_rmse:.4f})")

# Plot the complete model and the original data
model_aic =  AIC(8, predicted, BES1[:, 1])
ax.plot(vT, vS, color = "coral", lw = 3, label = f"Predicted BES1 (RMSE = {rmse:.4f})")
ax.scatter(BES1[:, 0], BES1[:, 1], color = "k", alpha = 0.4, edgecolor = "None")

# Create a table which shows the RMSE and AIC
axins = ax.inset_axes([0.05, 0.35, 0.5, 0.6])
axins.axis("off")
axins.tick_params(which = "both", left = False, bottom = False, labelleft = False, labelbottom = False)

table_text = [["Model", "RMSE", r"$\Delta$ AIC"], 
              ["Linear", f"{linear_rmse:.4f}", f"{linear_aic + 305:.4f}"],
              ["Exponential", f"{exp_rmse:.4f}", f"{exp_aic + 305:.4f}"], 
              ["Complete", f"{rmse:.4f}", f"{model_aic + 305:.4f}"]]
table_colours = [["white", "white", "white"],
                 ["limegreen", "white", "white"],
                 ["slateblue", "white", "white"],
                 ["coral", "white", "white"]]

table = axins.table(cellText = table_text, cellColours = table_colours, loc = "upper left")
table.scale(1, 2.5)

fig.savefig(f"{PATH}/bes1-model-fit.png", bbox_inches = "tight")
```
