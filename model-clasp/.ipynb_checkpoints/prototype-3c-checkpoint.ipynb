```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import minimize, Bounds
from sympy import symbols, solve, lambdify
```

```python
mpl.rcParams['font.size'] = 22
mpl.rcParams['figure.figsize'] = (12, 8)
```

```python
# Start the testing data at 150um from the QC, since this is when the last division takes place
TIMES =  np.array([6.7, 11, 15, 18, 20, 21, 22])
PRECURSORS = np.array([1.5, 1.65, 2, 2.35, 2.85, 2.65, 2.65])
SIGNALS = np.array([1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5])
SIZES = np.array([7, 9, 13, 20, 30, 42, 55])
```

```python
# Find the polynomial of best fit and graph it
RB = np.poly1d(np.polyfit(TIMES, SIGNALS, 2))
xs = np.linspace(min(TIMES), max(TIMES), 100)
plt.plot(xs, RB(xs), color = "blue", ls = "--", label = r"$R_B(t)$")

# Add a scatterplot of the original points
plt.scatter(TIMES, SIGNALS, color = "red", label = "Experimental Data")
plt.xlabel("Time (h)")
plt.ylabel("BR Signalling")
plt.legend()
plt.savefig("img/prototype-3-rb.png")
```

## Equations

$$
\begin{cases}
\dfrac{ dC }{ dt } = c_{\text{max}} - (1 + qR_{B})C \\[5pt]
\dfrac{ dM }{ dt } = \alpha _{M }C - M \\[5pt]
\dfrac{ dL }{ dt } = \dfrac{ g_{B}R_{B} }{ 1 + (g_{M}M / L) }
\end{cases}
$$

```python
START = 6.7
STOP = 22
STEP = 0.05
LENGTH = 7
SERIES = np.arange(START, STOP + STEP, STEP)
INITIAL_PARAMS = np.ones(5)
```

```python
def setup():

    # Define parameters and variables
    params = symbols('c_max, q, a_M, g_B, g_M')
    c_max, q, a_M, g_B, g_M = params
    
    # Define the initial cell length and symbols
    C, RB, M = symbols('C RB M')
            
    # Solve the system at steady state
    system = [
        c_max - (1 + q*RB)*C,
        a_M*C - M
    ]

    # Choose the second root to ensure positive solutions
    steady_states = solve(system, [C, M], dict = True)[-1]
    return [lambdify([params, RB], steady_states[hormone]) for hormone in [C, M]]

HORMONES = setup()
```

```python
def get_growth(params): 
    c_max, q, a_M, g_B, g_M = params
    return lambda M, L, RB : (g_B * RB) / (1 + (g_M * M / L))

# Compute the NRMSE of the model
def check_error(sizes):
    indices = [int((t - START) / STEP) for t in TIMES]
    sizes = [sizes[i] for i in indices]
    return sum([(s - S) ** 2 for s, S in zip(sizes, SIZES)]) / sum([S ** 2 for S in SIZES])
    
def cost(params):
    
    sizes, L, growth = [], LENGTH, get_growth(params)
    C, M = [h(params, RB(SERIES)) for h in HORMONES]
    
    for m, rb in zip(M, RB(SERIES)):
        L += growth(m, L, rb) * STEP
        sizes.append(L)
    
    return check_error(sizes)

def simulate_cell(params):
    
    sizes, L, growth = [], LENGTH, get_growth(params)
    C, M = [h(params, RB(SERIES)) for h in HORMONES]
    
    for m, rb in zip(M, RB(SERIES)):
        L += growth(m, L, rb) * STEP
        sizes.append(L)
    
    return (C, M, sizes), check_error(sizes)
```

```python
%%time
res = minimize(
    cost, 
    INITIAL_PARAMS, 
    method='trust-constr', 
    bounds=Bounds(1, 100), 
    options={"maxiter": 10000}
)
```

```python
# Get the data for the initial parameters
data, error = simulate_cell(res.x)
C, M, sizes = data

# Plot the results
plt.plot(SERIES, sizes, color = "green", label = "Fitted Model")
plt.scatter(TIMES, SIZES, color = "black", label = "Experimental Data")
plt.ylabel("Cell Length (um)")
plt.legend()
plt.xlabel("Time (h)")
plt.savefig("img/prototype-3c-model.png")

# Print off information about the optimization
print("SUCCESS:", res.success, res.message)
print("ITERATIONS:", res.nit)
print("PARAMS:", [round(p, 3) for p in list(res.x)])
print("ERROR:", error)
```
