```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
### Read csv files from /data

# Cell file data (time vs. position)
line1_raw = pd.read_csv('data/line1.csv', header = None)
line2_raw = pd.read_csv('data/line2.csv', header = None)
line3_raw = pd.read_csv('data/line3.csv', header = None)

# Cell length data (position vs. length)
sizes_raw = pd.read_csv('data/sizes.csv', header = None)

# Hormone signalling data (au vs. position)
bes1_raw = pd.read_csv('data/bes1.csv', header = None)
cpd_raw = pd.read_csv('data/cpd.csv', header = None)
rot3_raw = pd.read_csv('data/rot3.csv', header = None)

# Auxin data (au vs. cell number)
auxin_raw = pd.read_csv('data/auxin.csv', header = None)
```

```python
# Convert a column of a series to a numpy array
as_numpy = lambda df, idx : pd.Series(df[idx]).to_numpy()
```

```python
# Process cell file data
def get_file_data(df):

    # Get the sorted time series and position data
    ts = np.sort(as_numpy(df, 0))
    ps = np.sort(as_numpy(df, 1))

    # Shift the time values so that the first value is at t = 0h
    ts = ts - ts[0]

    # Shift the position values so that the first value is at l = 150um
    ps = ps - ps[0] + 150

    # Return the modified time and position series
    return ts, ps

# Get the cell file data from each file
ts_1, ps_1 = get_file_data(line1_raw)
ts_2, ps_2 = get_file_data(line2_raw)
ts_3, ps_3 = get_file_data(line3_raw)

# Concatenate the ts and ls together and sort them
ts = np.sort(np.concatenate((ts_1, ts_2, ts_3)))
ps = np.sort(np.concatenate((ps_1, ps_2, ps_3)))

# Combine data into a 2D numpy array
time_vs_position = np.column_stack((ts, ps))

# Plot the prepared data
plt.scatter(time_vs_position[:, 0], time_vs_position[:, 1])
```

```python
# Prepare any other data set and sort it by the first argument
def prepare_data(df):

    # Convert the columns to numpy arrays
    s0 = as_numpy(df, 0)
    s1 = as_numpy(df, 1)

    # Create a n x 2 numpy array and sort by the first argument
    arr = np.column_stack((s0, s1))
    return arr[arr[:, 0].argsort()]
```

```python
# Prepare and plot the sizes data
position_vs_length = prepare_data(sizes_raw)
plt.scatter(position_vs_length[:, 0], position_vs_length[:, 1])
```

```python
# Prepare and plot the BES1 data
position_vs_bes1 = prepare_data(bes1_raw)
print(position_vs_bes1[:, 0])
plt.scatter(position_vs_bes1[:, 0], position_vs_bes1[:, 1])
```

```python
# Prepare and plot the CPD data
position_vs_cpd = prepare_data(cpd_raw)
plt.scatter(position_vs_cpd[:, 0], position_vs_cpd[:, 1])
```

```python
# Prepare and plot the ROT3 data
position_vs_rot3 = prepare_data(rot3_raw)
plt.scatter(position_vs_rot3[:, 0], position_vs_rot3[:, 1])
```

```python
# Prepare and plot the auxin data
number_vs_auxin = prepare_data(auxin_raw)
plt.scatter(number_vs_auxin[:, 0], number_vs_auxin[:, 1])
```

```python
# Export data to JSON
import json
with open('data.json', 'w', encoding='utf-8') as f:
    data = {
        "POSITION": time_vs_position.tolist(),
        "LENGTH": position_vs_length.tolist(),
        "BES1": position_vs_bes1.tolist(),
        "CPD": position_vs_cpd.tolist(), 
        "ROT3": position_vs_rot3.tolist(), 
        "AUXIN": number_vs_auxin.tolist(),
    }
    json.dump(data, f, ensure_ascii=False, indent=4)
```
