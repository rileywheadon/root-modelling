```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import itertools
import json
import numpy as np

from matplotlib.gridspec import GridSpec
from numba import njit
from scipy.optimize import minimize, Bounds
from scipy.stats.qmc import Sobol
from sympy import symbols, solve, lambdify
from prototypefour import *
```

```python
with open("data.json") as f:
    POSITION, LENGTH, BES1, CPD, ROT3, AUXIN = map(np.array, json.load(f).values())

# Define time to position and position to time functions
time_to_position = np.poly1d(np.polyfit(POSITION[:, 0], POSITION[:, 1], 4))
position_to_time = np.poly1d(np.polyfit(POSITION[:, 1], POSITION[:, 0], 4))

# Define the time and vectors
STEP = 0.01
vT = np.arange(0, 18 + STEP, STEP)
vP = time_to_position(vT)

# Remove outliers from the BES1 data
idx = np.where(BES1[:, 1] < 100)
BES1 = BES1[idx]

# Filter the data to only include positions between 150 and 600um
# Transform the data to be in terms of time
# Also divide by 100 so that the numerics of the model are less erratic
def filter_transform(data):
    idx = np.where((data[:, 0] > 150) & (data[:, 0] < 600))
    data = data[idx]
    data[:, 0] = position_to_time(data[:, 0])
    data[:, 1] = data[:, 1] / 100
    return data

BES1 = filter_transform(BES1)
LENGTH = filter_transform(LENGTH)
```

```python
# Compute the RMSE of a trial run given a predicted BR signalling level vS
@njit
def RSS(vS):
    predicted = np.interp(BES1[:, 0], vT, vS)
    observed = BES1[:, 1]
    residuals = predicted - observed
    rss = np.sum(np.square(residuals))
    return rss / residuals.shape[0]

# Run a complete simulation
@njit
def simulate_cell(params):

    s_0, s_in, s_out = params
    vB = get_br(vP, CPD, ROT3)
    vS, s = np.array([np.float64(x) for x in range(0)]), s_0
    
    for b in vB:
        s = s + (s_in * b - s_out * s) * STEP
        vS = np.append(vS, s)
        
    return [vB, vS], RSS(vS)

# Define the cost function
@njit
def cost(params):
    return simulate_cell(params)[-1]
```

## Model Fitting

```python
params = np.array([0.5, 0.05, 0.05])
bounds = np.array([[0, 1], [0, 0.1], [0, 0.1]])

# Find the parameters of best fit
fit = minimize(
    cost, 
    params, 
    method = "trust-constr",
    bounds = bounds, 
    options = {"maxiter": 1000}
)

# Get the number of parameters
k = len(params) + 1

# Run a simulation with the optimal parameters
data, error = simulate_cell(fit.x)

# Log the simulation
print("Success: ", fit.success, fit.message)
print("Params: ", [round(n, 4) for n in fit.x])
print("Signal Error: ", round(error * 10000, 4))

# n = BES1.shape[0]
# AIC = 2 * k + n * np.log(error) + ((2 * k ** 2 + 2 * k) / (n - k - 1))
# print("AIC: ", AIC, "\n")
```

## Uncertainty Analysis

```python
mpl.rcParams['font.size'] = 22
mpl.rcParams['figure.figsize'] = (10, 10)
```

```python
# Compute the moving average of CPD, then interpolate it
sCPD_MA = np.interp(CPD[:, 0], np.linspace(0, 999, 1000), moving_average(CPD))

# Find the error between the moving average and the raw data
std = np.std(CPD[:, 1] - sCPD_MA)

# Get random modified input data
sCPD_R = np.random.normal(0, std, size = CPD.shape[0]) + CPD[:, 1]

# Get the randomized moving average and interpolate it
CPD_R = np.column_stack((CPD[:, 0], sCPD_R))
sCPD_MA_R = np.interp(CPD[:, 0], np.linspace(0, 999, 1000), moving_average(CPD_R))

# Plot the original data
# plt.scatter(CPD[:, 0], CPD[:, 1], marker=".")
plt.plot(CPD[:, 0], sCPD_MA, linewidth = 3)

# Plot the randomized data
# plt.scatter(CPD[:, 0], cpd_randomized, marker=".")
plt.plot(CPD[:, 0], sCPD_MA_R, linewidth = 3)
```
