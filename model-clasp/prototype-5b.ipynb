```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import itertools
import json
import numpy as np
import itertools

from IPython.display import clear_output
from numpy.polynomial import Polynomial
from matplotlib.gridspec import GridSpec
from numba import njit, prange
from scipy.optimize import minimize, Bounds
from scipy.stats.qmc import Sobol
from sympy import symbols, solve, lambdify
from prototypefive import *
```

```python
with open("data.json") as f:
    POSITION, LENGTH, BES1_RAW, CPD, ROT3, AUXIN = map(np.array, json.load(f).values())

# Define time to position and position to time functions
time_to_position = np.poly1d(np.polyfit(POSITION[:, 0], POSITION[:, 1], 4))
position_to_time = np.poly1d(np.polyfit(POSITION[:, 1], POSITION[:, 0], 4))

# Define the time and position vectors
STEP = 0.01
vT = np.arange(0, 18 + STEP, STEP)
vP = time_to_position(vT)

# Preprocess the BES1 data
def process_data(data):
    
    # Remove outliers
    idx = np.where(data[:, 1] < 100)
    data = data[idx]

    # Filter the data to only include positions between 150 and 600um
    idx = np.where((data[:, 0] > 150) & (data[:, 0] < 600))
    data = data[idx]

    # Normalize the BES1 data to have a maximum of 1
    data[:, 1] = data[:, 1] / np.max(data[:, 1])

    # Transform the BES1 data to be in terms of time
    data[:, 0] = position_to_time(data[:, 0])
    return data

BES1 = process_data(BES1_RAW)
```

```python
# Compute the RMSE of a trial run given a predicted BR signalling level vS
@njit
def RMSE(vS, data):

    # Compute the residuals
    predicted = np.interp(BES1[:, 0], vT, vS)
    observed = data
    residuals = predicted - observed

    # Find the residual sum of squares and RMSE
    rss = np.sum(np.square(residuals))
    rmse = np.sqrt(rss / np.size(data))
    return rmse, predicted

# Function for computing the bound receptors
@njit
def RB(B, R_T, K_D):
    A = B + R_T + K_D
    return (A - np.sqrt(A ** 2 - 4 * R_T * B)) / 2

# Run a complete simulation
@njit
def simulate_cell(data, params):

    s_0, s_in, s_out = params
    vB = get_br(vP, CPD, ROT3)
    vRB = RB(vB, 62, 10)
    s, vS = s_0, np.empty(vRB.size)
    
    for i in prange(vRB.size):
        vS[i] = s
        s += (s_in * vRB[i] - s_out * s) * STEP

    rmse, predicted = RMSE(vS, data)
    return vS, rmse, predicted
```

## Model Best Fit

```python
def fit_model(data, log):

    # Set initial parameter values and bounds
    params = np.array([0.25, 0.25, 0.25])
    bounds = np.array([[0, 0.5], [0, 0.5], [0, 0.5]])

    # Create a cost function
    cost = lambda params : simulate_cell(data, params)[1]

    # Find the parameters of best fit
    fit = minimize(
        cost, 
        params, 
        method = "trust-constr",
        bounds = bounds, 
        options = {"maxiter": 1000}
    )

    # Run a simulation with the optimal parameters
    vS, rmse, predicted = simulate_cell(BES1[:, 1], fit.x)

    # Log the simulation
    if log: 
        print("Success: ", fit.success, fit.message)
        print("Params: ", fit.x)
        print("Error: ", rmse)
        
    return vS, rmse, predicted, fit.x

vS, rmse, predicted, optimal_params = fit_model(BES1[:, 1], True)
```

```python
plt.plot(vT, vS, color = "orange", label = "Predicted BES1")
plt.scatter(BES1[:, 0], BES1[:, 1], label = "Observed BES1")
plt.xlabel(r"Time ($h$)")
plt.ylabel("BES1 Transcription Factor")
plt.xticks(np.arange(0, 20, 2))
plt.title(fr"Parameters: {[round(x, 4) for x in optimal_params]}, RMSE: {rmse:.4f}")
plt.legend()

mpl.rcParams['font.size'] = 18
mpl.rcParams['figure.figsize'] = (10, 8)
plt.savefig("img/prototype-5b-best-fit.png", bbox_inches = "tight")
```
