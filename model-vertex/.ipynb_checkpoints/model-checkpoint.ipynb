```python
import matplotlib.pyplot as plt
import matplotlib.patches as patch
import numpy as np
```

```python
fig, ax = plt.subplots(figsize = (12, 12))
ax.set_aspect('equal')
ax.set_xlim(-150, 150)
ax.set_ylim(0, 900)
ax.invert_yaxis()

W, C = generate_network()
for c in C:
    ax.add_patch(c.view())
    
for w in W:
    x, y = w.view()
    ax.plot(x, y, 'k-', lw = 1)
    
#for w in W:
```
