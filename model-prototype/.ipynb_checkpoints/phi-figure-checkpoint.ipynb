```python
from computation.gradient import get_phi, get_clasp
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Import configuration
import json
with open("config.json") as f:
    CONFIG = json.load(f)
    PARAMS = CONFIG["PARAMS"]
    MODEL = CONFIG["MODEL"]
```

```python
# Set Matplotlib display settings
import matplotlib.pyplot as plt
plt.rcParams['axes.titlepad'] = 20 
plt.rcParams['hatch.color'] = 'lightgrey'
plt.rcParams['hatch.linewidth'] = 0.2
plt.rcParams['font.size'] = 18
plt.rcParams['savefig.facecolor'] = 'white'
```

```python
phi = get_phi(PARAMS["phi"])
clasp = get_clasp(PARAMS["phi"])

xs = np.linspace(0, 1, 100)
plt.plot(xs, phi(xs), label = r'$\phi(B)$', color = "green")
plt.plot(xs, clasp(xs), label = r'$C(B)$', color = "red")
plt.legend()
plt.savefig("figures/phi-clasp.png")
```
